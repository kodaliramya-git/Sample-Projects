```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
print(df.isnull().sum())
```

```python
df['director']= df['director'].fillna('Unknown')
```

```python
df['country']=df['country'].fillna('Not Available')
```

```python
print(df.isnull().sum())
```

```python
df.shape
```

```python
# Drop Rows with Missing Ratings
df.dropna(subset=['rating'], inplace = True)
```

```python
df.shape
```

```python
df
```

```python
# Drop Duplicates and statical information
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.describe()
```

```python
df.info()
```

```python
# movies vs tv shows
```

```python
df['type'].value_counts()
```

```python
content_type = df['type'].value_counts()
plt.bar(content_type.index,content_type.values)
plt.title('Movies vs TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)

top_countries.plot(kind='bar', figsize=(10,5))

plt.title('Top 10 Countries by Shows')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(7,7))

plt.pie(
    top_countries,
    labels=top_countries.index,
    autopct='%1.2f%%'
)

plt.title('Top 10 Countries by Shows')

plt.show()
```

```python
df['release_year'].value_counts().sort_index(ascending=False)
```

```python
# Count shows/movies by release year

year_counts = df['release_year'].value_counts().sort_index()

plt.figure(figsize=(10,5))

plt.plot(year_counts.index, year_counts.values, marker='o')

# Titles and labels
plt.title('Shows Released by Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Shows')
plt.show()
```

```python
rating = df['rating'].value_counts().head(5)

plt.figure(figsize=(8,7))

plt.pie(
    rating,
    labels=rating.index,
    autopct='%1.2f%%'
)

plt.title('Top Rating in Netflix')

plt.show()
```

```python
releaseyear_oldest = df['release_year'].value_counts().sort_index() 

print(releaseyear_oldest)


```

```python
releaseyear_newest = df['release_year'].value_counts().sort_index(ascending = False) 

print(releaseyear_newest)
```

```python
Newest_movies = df.sort_values(by='release_year',ascending= False)
print(Newest_movies[['title','release_year']].head(10))
```

```python
Oldest_movies = df.sort_values(by='release_year')
print(Oldest_movies[['title','release_year']].head(10))
```

```python
top_genres = df['listed_in'].value_counts().head(10)
print(top_genres)
```

```python
plt.figure(figsize=(12,5))
plt.bar(top_genres.index,top_genres.values)
plt.xticks(rotation=75)

plt.title('Top Genres')
plt.xlabel('Genres')
plt.ylabel('Count')
plt.show()
```
